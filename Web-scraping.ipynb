```python
import requests
from bs4 import BeautifulSoup
import time

base_url = "https://www.flipkart.com/search?q=toys&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

all_products = []

for page in range(1, 5):  # Change 5 to 31 for 30 pages
    
    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all product containers on the page
    products = soup.find_all("div", class_="slAVV4")

    for product in products:
        title_tag = product.find("a", class_="wjcEIp")
        rating_tag = product.find("div", class_="XQDdHH")
        price_tag = product.find("div", class_="Nx9bqj")
        discount_tag = product.find("div", class_="UkUFwK")

        if title_tag:
            item = {
                "title": title_tag.text.strip(),
                "rating": rating_tag.text.strip() if rating_tag else "No rating",
                "price": price_tag.text.strip() if price_tag else "No price",
                "discount": discount_tag.text.strip() if discount_tag else "No discount"
            }
            all_products.append(item)
    
    time.sleep(1)  # polite delay between requests

# Print first 5 results to verify
for i, product in enumerate(all_products[:5], 1):
    print(f"\nProduct {i}:")
    for key, value in product.items():
        print(f"{key.capitalize()}: {value}")
```

```python
import requests
from bs4 import BeautifulSoup
import time
import csv

base_url = "https://www.flipkart.com/search?q=toys&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

all_products = []

for page in range(1, 10):  # Change 10 to 31 for 30 pages
    print(f"Scraping page {page}...")
    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    products = soup.find_all("div", class_="slAVV4")

    for product in products:
        title_tag = product.find("a", class_="wjcEIp")
        rating_tag = product.find("div", class_="XQDdHH")
        price_tag = product.find("div", class_="Nx9bqj")
        discount_tag = product.find("div", class_="UkUFwK")

        if title_tag:
            item = {
                "title": title_tag.text.strip(),
                "rating": rating_tag.text.strip() if rating_tag else "No rating",
                "price": price_tag.text.strip() if price_tag else "No price",
                "discount": discount_tag.text.strip() if discount_tag else "No discount"
            }
            all_products.append(item)
    
    time.sleep(1)  # Be polite and avoid hitting the server too fast

csv_file_path = r"C:\Users\mohan\Downloads\products.csv"
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["title", "rating", "price", "discount"])
    writer.writeheader()
    writer.writerows(all_products)

print(f"\nSaved {len(all_products)} products to {csv_file_path}")
```
